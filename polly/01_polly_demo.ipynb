# Dependências

```python
import os
import json

import boto3
```

# Constantes

```python
# Pasta local para salvar respostas das chamadas AWS
OUTPUT_FOLDER = 'respostas'
```

# Criar pasta local para respostas

```python
OUTPUT_FOLDER = "respostas"
os.makedirs(OUTPUT_FOLDER, exist_ok=True)
```

# Cliente AWS

```python
polly_client = boto3.client("polly")
```

# Teste do método Response Syntax

```python
response = polly_client.synthesize_speech(
            Text="Neste vídeo, vamos explorar como usar inteligência artificial generativa para criar conteúdo automatizado de forma eficiente e escalável.",
            OutputFormat="mp3",
            VoiceId="Camila",
            LanguageCode="pt-BR",
            Engine="neural" # 'standard'|'neural'|'long-form'|'generative'
        )
        
```

```python
# Vamos ver a estrutura geral do resposne exportanto o resultado em um arquivo JSON
with open(f"{OUTPUT_FOLDER}/response.json", 'w', encoding='utf-8') as json_file:
    json.dump(response, json_file, ensure_ascii=False, indent=4, default=str)
```

# Salvar áudio localmente

```python
with open(f"{OUTPUT_FOLDER}/speech.mp3", "wb") as file:
        file.write(response["AudioStream"].read())
```
